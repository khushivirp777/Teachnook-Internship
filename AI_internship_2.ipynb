```python
def get_outlier(df, col):
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  upper_limit = Q3 + 1.5 * IQR
  lower_limit = Q1 - 1.5 * IQR

  upper_limit_index = df[df[col] > upper_limit].index
  lower_limit_index = df[df[col] < lower_limit].index

  outlier_abovedf = df.loc[upper_limit_index]
  outlier_belowdf = df.loc[lower_limit_index]

  print("outlier_above", outlier_abovedf)
  print("outlier_below", outlier_belowdf)

  print(f'outlier_above_upper_limit: {len(outlier_abovedf)}')
  print(f'outlier_below_lower_limit: {len(outlier_belowdf)}')


  return upper_limit_index, lower_limit_index
```

```python
import numpy as np
import pandas as pd

def outlier_remover(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR

    # Print the limits and count of outliers before removal
    print(f"Q1 for {col}: {Q1}")
    print(f"Q3 for {col}: {Q3}")
    print(f"upper_limit for {col}: {upper_limit}")
    print(f"lower_limit for {col}: {lower_limit}")
    print(f'Outliers Above Upper limit Before: {df[df[col] > upper_limit].shape[0]}')
    print(f'Outliers Below Lower limit Before: {df[df[col] < lower_limit].shape[0]}\n')

    # Remove outliers
    df_outliers_removed = df[(df[col] <= upper_limit) & (df[col] >= lower_limit)]

    # Print the count of outliers after removal
    print(f'Outliers Above Upper limit After: {df_outliers_removed[df_outliers_removed[col] > upper_limit].shape[0]}')
    print(f'Outliers Below Lower limit After: {df_outliers_removed[df_outliers_removed[col] < lower_limit].shape[0]}\n\n')
    print(df_outliers_removed)
    return df_outliers_removed

# Create a function to remove outliers from all numerical columns
def remove_outliers_from_all_columns(df):
    numerical_cols = df.select_dtypes(include=np.number).columns
    for col in numerical_cols:
        df = outlier_remover(df, col)
    return df
```
