```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
data
```

```python
X = pd.DataFrame(data.data, columns=data.feature_names)
X
```

```python
X.info()
```

```python
X_corr = X.corr()
X_corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(30, 20))
sns.heatmap(X_corr, annot=True)
```

```python
X_corr.abs()
X_corr
```

```python
threshhold = 0.9
hig_correlation = np.where(X_corr>threshhold)
hig_correlation
```

```python
hsig_corelation = [(X_corr.columns[x], X_corr.columns[y])
                 for x,y in zip(*hig_correlation)
                 if x!=y and x<y]
hsig_corelation
```

```python
columns_to_drop = set()

for pair in hsig_corelation:
    columns_to_drop.add(pair[1])

columns_to_drop
```

```python
X = X.drop(columns_to_drop, axis=1)
X
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_scaled = sc.fit_transform(X)
X
```

```python
X.std()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for k in range(1,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))

plt.plot(range(1,11), wcss, marker = "o")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
clusters = kmeans.fit_predict(X_scaled)
```

```python
X['Clusters'] = clusters
X
```

```python
from sklearn.metrics import silhouette_score
```

```python
sc = silhouette_score(X_scaled, clusters)
sc
```
